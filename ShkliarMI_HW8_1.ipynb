````python
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "bigscience/bloom-7b1"

print("Загрузка модели...")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)

# Создание пайплайна для генерации текста
generator = pipeline("text-generation", model=model, tokenizer=tokenizer, max_length=256)
````

````python
# Task 1: Разговор между знаменитостью и вымышленным персонажем
# Вариант (a): только имена
prompt_a = """Elon Musk:
Harry Potter:"""

output_a = generator(prompt_a, max_length=200, num_return_sequences=1, do_sample=True)
print("Вариант (a) ответ:\n", output_a[0]['generated_text'])
````

````python
# Вариант (b): имена + дополнительная информация
prompt_b = """Elon Musk (известный миллиардер, основатель SpaceX и Tesla) встречает Гарри Поттера (молодой волшебник из Хогвартса, победивший Темного Лорда).
Илон Маск интересуется магией и хочет узнать, можно ли использовать волшебство для космических путешествий.

Elon Musk:
Harry Potter:"""

output_b = generator(prompt_b, max_length=200, num_return_sequences=1, do_sample=True)
print("\nВариант (b) ответ:\n", output_b[0]['generated_text'])
````

__Task 2b: (non-BLOOM):__ toxicity classification for [SetFit/toxic_conversations](https://huggingface.co/datasets/SetFit/toxic_conversations). Make the model solve binary classification (toxic vs not toxic) in the few shot mode. For few-shot examples, use 2-3 toxic and 2-3 non-toxic non-toxic examples. Measure accuracy on at least 25 samples. You may need to try several different prompts before you find the one that works.

````python
original_text = """Once upon a midnight dreary, while I pondered, weak and weary,
Over many a quaint and curious volume of forgotten lore—
While I nodded, nearly napping, suddenly there came a tapping,
As of some one gently rapping, rapping at my chamber door.
“’Tis some visitor,” I muttered, “tapping at my chamber door—
Only this and nothing more.”"""

# Пытаемся заставить модель перевести текст на французский
prompt_translate = f"Translate the following English text into French:\n\n{original_text}\n\nFrench Translation:"
translation_output = generator(prompt_translate, max_new_tokens=256, num_return_sequences=1, do_sample=True)[0]['generated_text']

# Извлекаем французский текст
import re
french_text_match = re.search(r'French Translation:\s*(.*)', translation_output, re.DOTALL)
french_text = french_text_match.group(1).strip() if french_text_match else translation_output.strip()
print("\nПеревод на французский:\n", french_text)

# Теперь пытаемся перевести обратно на английский
prompt_back = f"Translate the following French text into English:\n\n{french_text}\n\nEnglish Translation:"
back_translation_output = generator(
    prompt_back,
    max_new_tokens=256,
    num_return_sequences=1,
    do_sample=True
)[0]['generated_text']

# Извлекаем обратный перевод
english_text_match = re.search(r'English Translation:\s*(.*)', back_translation_output, re.DOTALL)
back_translated_text = english_text_match.group(1).strip() if english_text_match else back_translation_output.strip()
print("\nПеревод обратно на английский:\n", back_translated_text)
````


__Task 3 (1pt):__ create a prompt and few-shot examples tha make the model __change the gender pronouns__ of the main actor in a given sentence in any direction of your choice. E.g. the doctor took off _his_ mask <-> the doctor took of _her_ mask.


````python
# Few-shot примеры
few_shot_examples = """
Change the gender pronouns of the main actor in the following sentences:

1. The doctor took off his mask -> The doctor took off her mask.
2. The teacher explained her lesson -> The teacher explained his lesson.
3. The firefighter saved her cat -> The firefighter saved his cat.
"""

input_sentence = "The scientist presented his research."

# Создаем промпт
prompt = f"{few_shot_examples}\n4. {input_sentence} ->"

# Заставляем модель изменить местоимения, используя уже загруженный генератор
output = generator(prompt, max_new_tokens=50, num_return_sequences=1, do_sample=True)[0]['generated_text']

result = output.split("\n4.")[1].strip() if "\n4." in output else output.strip()
print("Измененное предложение:\n", result)
````

__Task 4 (1pt):__ write a prompt and supply examples such that the model would __convert imperial units to metric units__ (miles -> kilometers; mph -> kph). More specifically, the model should rewrite a given sentence and replace all imperial units with their metric equivalents. After it works with basic distances and speed, try to find complicated examples where it does *not* work.

Please note that 1 mile is not equal to 1 km :)

````python
few_shot_examples = """
Rewrite the following sentences by converting all imperial units to metric units. Use the correct conversion rates (1 mile = 1.609 km, 1 mph = 1.609 kph):

1. The car was traveling at 60 mph -> The car was traveling at 96.56 kph.
2. The runner completed a 5-mile race -> The runner completed an 8.05-kilometer race.
3. The airplane was flying at 500 mph -> The airplane was flying at 804.67 kph.
4. They drove 10 miles to reach the city -> They drove 16.09 kilometers to reach the city.
"""

input_sentence = "The road trip covered 120 miles at an average speed of 65 mph."

prompt = f"{few_shot_examples}\n5. {input_sentence} ->"

output = generator(prompt, max_new_tokens=100, num_return_sequences=1, do_sample=True)[0]['generated_text']

result = output.split("\n5.")[1].strip() if "\n5." in output else output.strip()
print("Конвертированное предложение:\n", result)
````
